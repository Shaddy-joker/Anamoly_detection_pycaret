```python
# importing anomaly dataset from pycaret
from pycaret.datasets import get_data
data = get_data('anomaly')
data.shape
```

```python
# importing the required modules for anomaly detection
from pycaret.anomaly import * # import all modules for anomaly detection in pycaret
#setup the modules
exp = setup(data,session_id=2202326)
```

```python
# create a model
knn = create_model('knn')
```

```python
# plot the created model
plot_model(knn)
```

```python
# generate predections using trained model
predict = predict_model(knn,data=data)
predict
```

```python
results = assign_model(knn)
results
```

```python

from pycaret.classification import *
cls = setup(data=results, target='Anomaly',session_id=2202326)
```

```python
# return the best model for classification
best_model = compare_models(sort='Recall',)
```

```python
# prints the best model for classification of data
print(best_model)
```

```python
evaluate_model(best_model)
```

```python
# plotting functions
plot_model(best_model,plot='auc')
```

```python
plot_model(best_model,plot='confusion_matrix')
```
